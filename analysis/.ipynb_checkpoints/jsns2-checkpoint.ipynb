```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import uproot
import h5py
```

```python
energy = 1
ds1 = pd.read_csv("mc_data_1102/mcEvent_jsns2_"+str(energy)+"MeV_0.csv")
ds2 = pd.read_csv("mc_data_1102/mcPMT_jsns2_"+str(energy)+"MeV_0.csv")

for i in range(100):
    path1 = "mc_data_1102/mcEvent_jsns2_"+str(energy)+"MeV_"+str(i+1)+".csv"
    path2 = "mc_data_1102/mcPMT_jsns2_"+str(energy)+"MeV_"+str(i+1)+".csv"

    if os.path.isfile(path1) == False:
        continue
    else:
        

        ds_E = pd.read_csv(path1)
        ds_P = pd.read_csv(path2)

        ds1 = pd.concat([ds1,ds_E])
        ds2 = pd.concat([ds2,ds_P])
    print(i)
```
